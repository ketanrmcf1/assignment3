```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.metrics import roc_curve, auc, accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.model_selection import RandomizedSearchCV
np.random.seed(11)
```

```python
df=pd.read_csv("sonar.all-data",header=None)
df.head()
#Label encoding the ouput variable
label_encoder = preprocessing.LabelEncoder() 
```

```python
#Mixing the dataset
df=df.sample(frac=1,random_state=13)
#df[60]
```

```python
#Label Encoding : R=1 , M=0
df.columns
X=df.drop(60,axis=1)
df[60]= label_encoder.fit_transform(df[60])
y=df[60]
#print(X)
#print(y)
```

```python
# Train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc = RandomForestClassifier(random_state=13)
rfc.fit(X_train,y_train)

# predictions
rfc_predict = rfc.predict(X_test)
print(accuracy_score(y_test, rfc_predict))
```

```python
#Cross validation score
rfc_cv_score = cross_val_score(rfc, X, y, cv=5)
print(rfc_cv_score.mean())
```

```python
print("1. Confusion Matrix ")
print(confusion_matrix(y_test, rfc_predict))
print('\n')

print("2. Classification Report ")
print(classification_report(y_test, rfc_predict))
print('\n')
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 50, num = 2)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5,50,num = 2)]
#max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2 ,4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Creating the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
from sklearn.ensemble import RandomForestClassifier# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1)

# Fit the random search model
rf_random.fit(X_train,y_train)
print("")
```

```python
print("Optimum parameters: \n")
print(rf_random.best_params_)
```

```python
best_random = rf_random.best_estimator_
y_pred=best_random.predict(X_test)
y_pred1=best_random.predict(X_train)
# print(y_pred)
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("ROC AUC Score: ",roc_auc)
```

```python
y_pred=[1 if i >=0.5 else 0 for i in y_pred]
y_pred1=[1 if i >=0.5 else 0 for i in y_pred1]
print(y_pred)
print(y_pred1)
print("Test Accuracy score: ",accuracy_score(y_test, y_pred))
print("Train Accuracy score: ",accuracy_score(y_train, y_pred1))
```
